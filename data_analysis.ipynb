```python
def to_numpy(hdf5_dset: h5py.Dataset):
    array = np.empty(hdf5_dset.shape, hdf5_dset.dtype)
    hdf5_dset.read_direct(array)
    return array
```

```python
Q = to_numpy(f['Q'])
```
